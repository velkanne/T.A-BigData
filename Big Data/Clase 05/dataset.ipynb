```python
import sqlite3
import pandas as pd
import json

print(" Entorno preparado para el analisis de infraestructuras ")
```

```python
conn = sqlite3.connect("cerveceria_tp.db")
cursor = conn.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS PRODUCCION (
        id_lote INTEGER PRIMARY KEY,
        estilo TEXT NOT NULL,
        litros INTEGER,
        temperatura_fermentacion REAL CHECK(temperatura_fermentacion BETWEEN 0 AND 30)
    )
""")
```

```python
try:
    cursor.execute("INSERT INTO PRODUCCION VALUES (1, 'IPA', 500, 18.5)")
    conn.commit()
    print("SQL: Lote guardado correctamente")
except Exception as e:
    print(f"Error SQL:{e}")
```

```python
try:
    cursor.execute("INSERT INTO PRODUCCION VALUES (2, 'Stout', 300, 45.0)")
    conn.commit()
except Exception as e:
    print(f"Bloqueo SQL esperado:{e} (La restriccion CHECK funciono)")
```

```python
base_nosql = []

doc1 = {
    "id": 1,
    "estilo": "IPA",
    "detalles": {"temperatura": 18.5, "operario": "Francisco"}
}

doc2 = {
    "id": 2,
    "estilo": "Stout",
    "iot_sensor_data": [18.2, 18.5, 19.0, 18,4],
    "clima_exterior": "Lluvia en Osorno",
    "alertas": "Nivel de espuma alto",
    "maestro_cervecero": "Analista TP"
}

base_nosql.append(doc1)
base_nosql.append(doc2)


print(f" NoSQL: Se guardaron {len(base_nosql)} documentos con esquemas totalmente distintos.")
print(json.dumps(base_nosql, indent=2))
```

```python
print ("VISTA DE TABLA SQL (Estructura Fija):")
df_sql = pd.read_sql_query("SELECT * FROM PRODUCCION", conn)
display(df_sql)

print("\n" + "=" + "\n")
print("VISTA DE DOCUMENTOS NOSQL (Estructura Fija):")
for doc in base_nosql:
    print(f"\nID Lote: {doc["id"]} | Estilo: {doc["estilo"]}")
    for clave, valor in doc.items():
        if clave not in ["id", "estilo"]:
            print(f" -{clave.upper()}: {valor}")
```
