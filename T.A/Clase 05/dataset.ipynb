```python
import pandas as pd 
import numpy as np

data = {
    "ID": [101, 102, 103, 104, 105],
    "Sede": ["Valdivia","valdivia", "LA UNION", "Panguipulli", " Valdivia"],
    "Monto_Beca": [250000, 250000, np.nan, -50000, 200000],
    "Correo": ["juan@cft.cl", "ana@cft.cl", "pedro.cl", "lucho@cft.cl", "ana@cft.cl"]
}

df = pd.DataFrame(data)
```

```python
df = df.drop_duplicates()
```

```python
df["Monto_Beca"] = df["Monto_Beca"].fillna(df["Monto_Beca"].mean())
```

```python
df["Sede"] = df["Sede"].str.title()
```

```python
df.loc[df["Monto_Beca"]< 0, "Monto_Beca"] = 0
```
