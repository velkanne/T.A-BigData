```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix1.csv", sep=";")
df["type_num"] = df["type"].map({"Movie": 0, "TV Show": 1})

df["release_year"] = pd.to_numeric(df["release_year"], errors="coerce")

df = df.dropna(subset=["release_year", "type_num"])

x = df[["release_year"]]
y = df["type_num"]

modelo = DecisionTreeClassifier(criterion="gini", max_depth=2)
modelo.fit(x, y)

plt.figure(figsize=(10, 6))
plot_tree(modelo,
    filled=True,
    feature_names=["release_year"],
    class_names=["Movie", "TV Show"])
plt.title("Arbol de Decision: Clasificacion de Contenido")
plt.show()
```
