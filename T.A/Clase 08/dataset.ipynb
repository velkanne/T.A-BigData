```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("VISTA_TOTAL.xlsx")

```

```python
x = df[["PRECIO_COMPRA", "PRECIO_VENTA", "CANTIDAD"]]
y = df["Region_Vendedor"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=42)
```

```python
modelo_over = DecisionTreeClassifier(criterion="gini")
modelo_over.fit(x_train, y_train)
```

```python
modelo_poda = DecisionTreeClassifier(criterion="gini", max_depth=3)
modelo_poda.fit(x_train, y_train)
```

```python
plt.figure(figsize=(15,10))
plot_tree(modelo_poda, filled=True, feature_names=x.columns,
class_names=y.unique())
plt.show()
```
