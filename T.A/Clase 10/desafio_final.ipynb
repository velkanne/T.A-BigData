```python
import joblib
import numpy as np

modelo = joblib.load('cerebro_tienda_2026.pkl')
print('Modelo cargado exitosamente')
print('Clases:', modelo.classes_)
```

```python
print('=========================================')
print('   MOTOR DE DECISIONES - VIRTUALSUR')
print('=========================================')
print()

nombre = input('Nombre del cliente: ')
edad = int(input('Edad: '))
ingreso = int(input('Ingreso mensual: '))
tech = int(input('Nivel tecnologico (1-10): '))
touch = int(input('Necesidad de tocar el producto (1-10): '))
confianza = int(input('Confianza en pagos online (1-10): '))
```

```python
datos_cliente = [[edad, ingreso, tech, touch, confianza]]

prediccion = modelo.predict(datos_cliente)[0]
probabilidades = modelo.predict_proba(datos_cliente)[0]
confianza_pred = max(probabilidades) * 100

print(f'\n=========================================')
print(f'   RESULTADO PARA: {nombre.upper()}')
print(f'=========================================')
print(f'Categoria predicha: {prediccion}')
print(f'Confianza: {confianza_pred:.1f}%')
print()
print('Probabilidades por categoria:')
for clase, prob in zip(modelo.classes_, probabilidades):
    barra = '#' * int(prob * 30)
    print(f'  {clase:<8} {prob*100:5.1f}% |{barra}')
```

```python
print(f'\n=========================================')
print(f'   ESTRATEGIA DE VENTAS')
print(f'=========================================')

if prediccion == 'Store':
    print(f'\nCliente: {nombre}')
    print('Perfil: COMPRADOR DE TIENDA FISICA')
    print()
    print('Acciones para el vendedor:')
    print('  - Preparar demostracion del producto en tienda')
    print('  - Ofrecer atencion personalizada y prueba de producto')
    print('  - Activar programa de fidelizacion con tarjeta fisica')
    print('  - Sugerir productos complementarios durante la visita')
    print('  - Entregar cupon de descuento para proxima compra presencial')

elif prediccion == 'Online':
    print(f'\nCliente: {nombre}')
    print('Perfil: COMPRADOR DIGITAL')
    print()
    print('Promocion digital a enviar:')
    print('  - Enviar codigo de descuento 15% por correo electronico')
    print('  - Activar notificaciones push con ofertas personalizadas')
    print('  - Ofrecer envio gratuito en su primera compra online')
    print('  - Incluir en campana de retargeting en redes sociales')
    print('  - Habilitar chat en vivo para asesoria de compra remota')

elif prediccion == 'Hybrid':
    print(f'\nCliente: {nombre}')
    print('Perfil: COMPRADOR HIBRIDO')
    print()
    print('Facilidades logisticas a ofrecer:')
    print('  - Activar opcion Click & Collect (compra online, retira en tienda)')
    print('  - Permitir devolucion en tienda de compras hechas online')
    print('  - Ofrecer reserva online con prueba presencial')
    print('  - Sincronizar carrito digital con asistencia en sucursal')
    print('  - Enviar catalogo digital con disponibilidad en tienda cercana')

if confianza_pred < 60:
    print()
    print('ALERTA: Confianza baja ({:.1f}%). Se recomienda'.format(confianza_pred))
    print('interaccion humana directa para confirmar preferencia real.')
```
